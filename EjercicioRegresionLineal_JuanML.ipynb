Realiza los siguientes pasos usando el dataset `diabetes` de sklearn, aplica el modelo de regresi贸n lineal:
- Carga el dataset incluido en `sklearn`


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

diabetes = load_diabetes()

df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
#df.to_csv("Diabete_excel.csv", index=False)
df
#print(diabetes.DESCR)
```


- Selecciona una caracter铆stica, como por ejemplo el BMI

```python
X = df[['bmi']]
#Y = df['s2']
#X = diabetes.data
Y = diabetes.target
print(X)
print("\n", Y)
```


- Divide los datos en datos de testing y entrenamiento.

```python
#df.plot.scatter(x='s2', y='target')
# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
print(f"X_train: {X_train}")
print(f"X_test: {X_test}")
print(f"Y_train: {Y_train}")
print(f"Y_test: {Y_test}")
```

A continuaci贸n normalizamos el target

```python
scaler = StandardScaler()
# En este caso no normalizamos los datos "X", ya que partimos del hecho de que estos ya fueron normalizados
#X_train = scaler.fit_transform(X_train)
#X_test = scaler.transform(X_test)

Y_train = Y_train.reshape(-1,1)
Y_test = Y_test.reshape(-1,1)
Y_train = scaler.fit_transform(Y_train)
Y_test = scaler.transform(Y_test)
```


- Entrena el modelito

```python
# Crear el modelo de regresi贸n lineal
model = LinearRegression() # Esta funcion crea un objeto modelo de regresion lineal

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, Y_train) # Ajusta el modelo a los datos de entrenamiento, encontrando los mejores valores para los coeficientes e intercepto de la ecuaci贸n de la regresi贸n.
```


- Eval煤a el modelo usando el error cuadr谩tico medio.

```python
y_pred = model.predict(X_test)
#y_pred

mse = mean_squared_error(Y_test, y_pred)
print(f"Error cuadr谩tico medio:   {mse}")

r2 = r2_score(Y_test, y_pred)
print(f"R^2: {r2}")
```


- Visualiza la gr谩fica con la relaci贸n entre la caracter铆stica y el valor objetivo junto con las predicciones del modelo C:



```python
ax = plt.axes()
ax.scatter(X_train, Y_train, color='blue', label='Datos de entrenamiento')
ax.scatter(X_test, Y_test, color='red', label='Datos de prueba')
ax.scatter(X_test, y_pred, color='purple', label='Predicciones')
plt.plot(X_train, model.predict(X_train), color='green', label='L铆nea de Regresi贸n')
plt.legend()
plt.show()
```

A continuaci贸n, quitamos los datos de entrenamiento para que no distraigan de las predicciones

```python
ax = plt.axes()
ax.scatter(X_test, Y_test, color='red', label='Datos de prueba')
ax.scatter(X_test, y_pred, color='purple', label='Predicciones')
plt.plot(X_train, model.predict(X_train), color='green', label='L铆nea de Regresi贸n')
plt.legend()
plt.show()
```
